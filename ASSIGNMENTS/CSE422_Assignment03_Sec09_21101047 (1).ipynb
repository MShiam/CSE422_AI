```python


import random
Get_value=input("enter: ")
set_branch=int(Get_value[2])
set_depth=int(Get_value[0])*2
leaves_counter = set_branch**set_depth
global set_branch
global set_depth

alpha = float('-inf')
beta = float('inf')

get_range=input('Enter  minimum and maximum hp ').split()
mean_hp,highest_hp=int(get_range[0]),int(get_range[1])
array = [random.randint(mean_hp,highest_hp) for k in range(leaves_counter)]
leafs=""
for i in range(len(array)):
    if array[i]!=array[-1]:
        leafs+=str(array[i])+","
leafs+=str(array[-1])

    


       
def max_min(array,d,check,alpha,beta,idx,pruned_list):
    if d == set_depth:
        return array[idx]
    if check:
        a = alpha
        for i in range(set_branch):
            val = max_min(array, d + 1, False, alpha, beta, idx * 2 + i, pruned_list)
            a = max(a, val)
            alpha = max(alpha, a)
            if alpha >= beta:
                pruned_list.append(array[idx * 2 + i])
                break
        return a
    else:
        b = beta
        for i in range(set_branch):
            val = max_min(array, d + 1, True, alpha, beta, idx * 2 + i, pruned_list)
            b = min(b, val)
            beta = min(beta, b)
            if alpha >= beta:
                pruned_list.append(array[idx * 2 + i])
                break
        return b

pruned_list=[]
answer = max_min(array, 0, True, alpha, beta, 0, pruned_list)
print(f"Depth and Branches ratio is {set_depth}:{set_branch}")
print("Terminal States (leaf node values) are",leafs)
print("Left life(HP) of the defender after maximum damage caused by the attacker is:",answer)
#print("Nodes pruned during alpha-beta pruning:", pruned_list)
remaining_nodes=len(array)-len(pruned_list)
print("After Alpha-Beta Pruning Leaf Node Comparisons ",remaining_nodes)

```
